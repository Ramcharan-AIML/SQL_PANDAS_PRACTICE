```python
# SECTION A — INNER JOIN (10 questions)

# 1.  INNER JOIN employees and departments
#     Get: emp_id, name, salary, dept_name
query_1 = '''
    SELECT 
        e.emp_id,
        e.name,
        e.salary,
        d.dept_name
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
''' 
question_1 = pd.read_sql(query_1 , conn)
print(question_1)
print("\n" , "="*100 , "\'n")

    
# 2.  INNER JOIN employees and departments
#     Get: name, dept_name, location
#     Only employees from "Technology" department
#     Hint: add WHERE d.dept_name = 'Technology'
query_2 = ''' 
    SELECT 
        e.name,
        d.dept_name,
        d.location
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE d.dept_name = "Technology"
''' 
question_2 = pd.read_sql(query_2,conn)
print(question_2)
print("\n" , "="*100 , "\'n")


# 3.  INNER JOIN employees and departments
#     Get: dept_name, COUNT(*) as emp_count
#     How many employees per department?
#     Hint: GROUP BY d.dept_name
query_3 = ''' 
    SELECT 
        d.dept_name,
        COUNT(*) as emp_count
    FROM employees e
    INNER JOIN  departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
''' 
question_3 = pd.read_sql(query_3, conn)
print(question_3)
print("\n" , "="*100 , "\'n")

# 4.  INNER JOIN employees and departments
#     Get: dept_name, AVG(salary) as avg_salary
#     Average salary per department
#     Round to 2 decimals
query_4 = ''' 
    SELECT 
        d.dept_name,
        ROUND(AVG(e.salary),2) as avg_salary
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
''' 
question_4 = pd.read_sql(query_4 , conn)
print(question_4)
print("\n" , "="*100 , "\'n")

# 5.  INNER JOIN employees and departments
#     Get: dept_name, MAX(salary), MIN(salary)
#     Max and min salary per department
query_5 = ''' 
    SELECT 
        d.dept_name,
        MAX(e.salary) as Maxium_salary,
        MIN(e.salary) as Minimum_salary
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
''' 
question_5 = pd.read_sql(query_5 , conn)
print(question_5)
print("\n" , "="*100 , "\'n")

# 6.  INNER JOIN employees and departments
#     Get all columns from both tables
#     Only employees whose salary > 60000
#     ORDER BY salary DESC
query_6 = ''' 
    SELECT 
        *
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE e.salary > 60000
    ORDER BY e.salary DESC
''' 
question_6 = pd.read_sql(query_6 , conn)
print(question_6)
print("\n" , "="*100 , "\'n")

# 7.  INNER JOIN employees and departments
#     Get: name, salary, dept_name, manager
#     Who is each employee's manager?
query_7 = '''
    SELECT 
        e.name,
        e.salary,
        d.dept_name,
        d.manager
    FROM  employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
''' 
question_7 = pd.read_sql(query_7 , conn)
print(question_7)
print("\n" , "="*100 , "\'n")

# 8.  INNER JOIN employees and departments
#     Get: dept_name, SUM(salary) as total_cost
#     Total salary cost per department
#     ORDER BY total_cost DESC
query_8 = ''' 
    SELECT 
        d.dept_name ,
        SUM(e.salary) as total_cost
    FROM employees e
    INNER JOIN departments d
        ON  e.dept_id = d.dept_id
    GROUP BY d.dept_name
    ORDER BY total_cost DESC
''' 
question_8 = pd.read_sql(query_8 ,conn)
print(question_8)
print("\n" , "="*100 , "\'n")

# 9.  INNER JOIN employees and departments
#     Get: name, city, dept_name
#     Only employees where employee city
#     matches department location
#     Hint: WHERE e.city = d.location
query_9 = ''' 
    SELECT
        e.name,
        e.city,
        d.dept_name
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE e.city = d.location
''' 
question_9 = pd.read_sql(query_9 , conn)
print(question_9)
print("\n" , "="*100 , "\'n")


# 10. INNER JOIN employees and salary_history
#     Get: name, year, salary from salary_history
#     All salary records with employee names
query_10 = ''' 
    SELECT 
        e.name,
        s.year,
        s.salary
    FROM employees e
    INNER JOIN salary_history S
        ON e.emp_id = s.emp_id
''' 
question_10 = pd.read_sql(query_10 , conn)
print(question_10)

# # -----------------------------------------------
```
