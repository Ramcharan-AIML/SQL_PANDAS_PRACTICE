```python
# SECTION A — INNER JOIN (10 questions)

# 1.  INNER JOIN employees and departments
#     Get: emp_id, name, salary, dept_name
query_1 = '''
    SELECT 
        e.emp_id,
        e.name,
        e.salary,
        d.dept_name
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
''' 
question_1 = pd.read_sql(query_1 , conn)
print(question_1)
print("\n" , "="*100 , "\'n")

    
# 2.  INNER JOIN employees and departments
#     Get: name, dept_name, location
#     Only employees from "Technology" department
#     Hint: add WHERE d.dept_name = 'Technology'
query_2 = ''' 
    SELECT 
        e.name,
        d.dept_name,
        d.location
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE d.dept_name = "Technology"
''' 
question_2 = pd.read_sql(query_2,conn)
print(question_2)
print("\n" , "="*100 , "\'n")


# 3.  INNER JOIN employees and departments
#     Get: dept_name, COUNT(*) as emp_count
#     How many employees per department?
#     Hint: GROUP BY d.dept_name
query_3 = ''' 
    SELECT 
        d.dept_name,
        COUNT(*) as emp_count
    FROM employees e
    INNER JOIN  departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
''' 
question_3 = pd.read_sql(query_3, conn)
print(question_3)
print("\n" , "="*100 , "\'n")

# 4.  INNER JOIN employees and departments
#     Get: dept_name, AVG(salary) as avg_salary
#     Average salary per department
#     Round to 2 decimals
query_4 = ''' 
    SELECT 
        d.dept_name,
        ROUND(AVG(e.salary),2) as avg_salary
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
''' 
question_4 = pd.read_sql(query_4 , conn)
print(question_4)
print("\n" , "="*100 , "\'n")

# 5.  INNER JOIN employees and departments
#     Get: dept_name, MAX(salary), MIN(salary)
#     Max and min salary per department
query_5 = ''' 
    SELECT 
        d.dept_name,
        MAX(e.salary) as Maxium_salary,
        MIN(e.salary) as Minimum_salary
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
''' 
question_5 = pd.read_sql(query_5 , conn)
print(question_5)
print("\n" , "="*100 , "\'n")

# 6.  INNER JOIN employees and departments
#     Get all columns from both tables
#     Only employees whose salary > 60000
#     ORDER BY salary DESC
query_6 = ''' 
    SELECT 
        *
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE e.salary > 60000
    ORDER BY e.salary DESC
''' 
question_6 = pd.read_sql(query_6 , conn)
print(question_6)
print("\n" , "="*100 , "\'n")

# 7.  INNER JOIN employees and departments
#     Get: name, salary, dept_name, manager
#     Who is each employee's manager?
query_7 = '''
    SELECT 
        e.name,
        e.salary,
        d.dept_name,
        d.manager
    FROM  employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
''' 
question_7 = pd.read_sql(query_7 , conn)
print(question_7)
print("\n" , "="*100 , "\'n")

# 8.  INNER JOIN employees and departments
#     Get: dept_name, SUM(salary) as total_cost
#     Total salary cost per department
#     ORDER BY total_cost DESC
query_8 = ''' 
    SELECT 
        d.dept_name ,
        SUM(e.salary) as total_cost
    FROM employees e
    INNER JOIN departments d
        ON  e.dept_id = d.dept_id
    GROUP BY d.dept_name
    ORDER BY total_cost DESC
''' 
question_8 = pd.read_sql(query_8 ,conn)
print(question_8)
print("\n" , "="*100 , "\'n")

# 9.  INNER JOIN employees and departments
#     Get: name, city, dept_name
#     Only employees where employee city
#     matches department location
#     Hint: WHERE e.city = d.location
query_9 = ''' 
    SELECT
        e.name,
        e.city,
        d.dept_name
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE e.city = d.location
''' 
question_9 = pd.read_sql(query_9 , conn)
print(question_9)
print("\n" , "="*100 , "\'n")


# 10. INNER JOIN employees and salary_history
#     Get: name, year, salary from salary_history
#     All salary records with employee names
query_10 = ''' 
    SELECT 
        e.name,
        s.year,
        s.salary
    FROM employees e
    INNER JOIN salary_history S
        ON e.emp_id = s.emp_id
''' 
question_10 = pd.read_sql(query_10 , conn)
print(question_10)

# # -----------------------------------------------
```

```python
# SECTION B — LEFT JOIN (10 questions)

# 11. LEFT JOIN employees and departments
#     Get ALL employees + their dept_name
#     (NULL if no department)
query_11 = ''' 
    SELECT 
        e.*,
        d.dept_name
    FROM employees e
    LEFT JOIN departments d
        ON e.dept_id = d.dept_id
''' 
question_11 = pd.read_sql(query_11 , conn)
print(question_11)
print("\n" , "="*100 , "\'n")

# 12. LEFT JOIN employees and departments
#     Find employees who have 
#     NO department assigned
#     Hint: WHERE d.dept_id IS NULL
query_12 = ''' 
    SELECT 
        e.*
    FROM employees e
    LEFT JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE d.dept_id IS NULL
''' 
question_12 = pd.read_sql(query_12 , conn)
print(question_12)
print("\n" , "="*100 , "\'n")

# 13. LEFT JOIN employees and departments
#     Count total employees per department
#     INCLUDING employees with no department
#     Show NULL dept as "No Department"
#     Hint: 
#     COALESCE(d.dept_name, 'No Department')
query_13 = ''' 
    SELECT 
        COALESCE(d.dept_name, 'No Department') as dept_name,
        COUNT(*) as total_employees
    FROM employees e
    LEFT JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY COALESCE(d.dept_name, 'No Department') as dept_name,
''' 
question_13 = pd.read_sql(query_13 , conn)
print(question_13)
print("\n" , "="*100 , "\'n")

# 14. LEFT JOIN employees and salary_history
#     Get ALL employees + their 2024 salary
#     Hint: WHERE s.year = 2024 OR s.year IS NULL
query_14 = ''' 
    SELECT
        *
    FROM employees e
    LEFT JOIN salary_history s
        ON e.emp_id = s.emp_id
    WHERE s.year = 2024 OR s.year IS NULL
''' 
question_14 = pd.read_sql(query_14 , conn)
print(question_14)
print("\n" , "="*100 , "\'n")

# 15. LEFT JOIN employees and salary_history
#     Find employees who have 
#     NO salary history record
#     Hint: WHERE s.emp_id IS NULL
query_15 = ''' 
    SELECT 
        *
    FROM employees e
    LEFT JOIN salary_history s
        ON e.emp_id = s.emp_id
    WHERE s.emp_id IS NULL
''' 
question_15 = pd.read_sql(query_15,conn)
print(question_15)
print("\n" , "="*100 , "\'n")

# 16. LEFT JOIN employees and departments
#     Get: name, salary, dept_name
#     Sort by salary DESC
#     Show all employees including 
#     those without departments
query_16 = ''' 
    SELECT 
        e.name,
        e.salary,
        d.dept_name
    FROM employees e
    LEFT JOIN departments d
        ON e.dept_id = d.dept_id
    ORDER BY e.salary DESC
''' 
question_16 = pd.read_sql(query_16,conn)
print(question_16)
print("\n" , "="*100 , "\'n")

# 17. LEFT JOIN employees and departments
#     Find departments that have 
#     MORE than 2 employees
#     Hint: GROUP BY + HAVING COUNT(*) > 2
query_17 = ''' 
    SELECT 
        d.dept_name
    FROM employees e
    LEFT JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
    HAVING COUNT(*) > 2
''' 
question_17 = pd.read_sql(query_17 , conn)
print(question_17)
print("\n" , "="*100 , "\'n")

# 18. LEFT JOIN employees and departments
#     Get average salary per department
#     Including employees with no department
#     Show as "Unassigned" for NULL dept
query_18 = ''' 
    SELECT 
        COALESCE(d.dept_name, "unassigned") as dept_name,
        AVG(e.salary) as avg_salary
    FROM employees e
    LEFT JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY COALESCE(d.dept_name, "unassigned") as dept_name,
''' 
question_18 = pd.read_sql(query_18 , conn)
print(question_18)
print("\n" , "="*100 , "\'n")


# 19. LEFT JOIN employees and salary_history
#     For each employee find their 
#     latest salary (MAX year)
#     Hint: GROUP BY e.emp_id
query_19 = ''' 
    SELECT 
        e.emp_id,
        s.salary,
        MAX(s.year)
    FROM employees e
    LEFT JOIN salary_history S
        ON e.emp_id = s.emp_id
    GROUP BY e.emp_id
''' 
question_19 = pd.read_sql(query_19,conn)
print(question_19)
print("\n" , "="*100 , "\'n")


# 20. LEFT JOIN all 3 tables together:
#     employees + departments + salary_history
#     Get: name, dept_name, year, bonus
#     All records including NULLs
query_20 = ''' 
    SELECT 
        e.name,
        d.dept_name,
        s.year,
        s.bonus
    FROM employees e
    LEFT JOIN departments d
        ON e.dept_id = d.dept_id
    LEFT JOIN salary_history s
        ON e.emp_id = s.emp_id
''' 
question_20 = pd.read_sql(query_20,conn)
print(question_20)
    
# # -----------------------------------------------
```

```python
# SECTION C — Advanced + Real World (10 questions)

# 21. INNER JOIN employees and departments
#     Find the department with 
#     highest average salary
#     Hint: GROUP BY + ORDER BY + LIMIT 1
query_21 = ''' 
    SELECT 
        d.dept_name,
        AVG(e.salary) as avg_sal
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
    ORDER BY avg_sal DESC
    LIMIT 1
''' 
question_21 = pd.read_sql(query_21 , conn)
print(question_21)
print("\n" , "="*100 , "\'n")


# 22. INNER JOIN employees and salary_history
#     Find total bonus received 
#     per employee across all years
query_22  ='''
    SELECT 
        e.name,
        s.year,
        SUM(s.bonus) as total_bonus
    FROM employees e
    INNER JOIN salary_history s
        ON e.emp_id = s.emp_id
    GROUP BY e.name
''' 
question_22 = pd.read_sql(query_22,conn)
print(question_22)
print("\n" , "="*100 , "\'n")

# 23. INNER JOIN employees and salary_history
#     Find employees whose salary 
#     increased from 2023 to 2024
#     Hint: Use two separate joins or subquery
#     This is a hard one — try your best
query_23 = ''' 
    SELECT s2023.emp_id, e.name,
       s2023.salary as salary_2023,
       s2024.salary as salary_2024
    FROM salary_history s2023
    INNER JOIN salary_history s2024
        ON s2023.emp_id = s2024.emp_id
    INNER JOIN employees e
        ON s2023.emp_id = e.emp_id
    WHERE s2023.year = 2023
      AND s2024.year = 2024
      AND s2024.salary > s2023.salary
'''
question_23 = pd.read_sql(query_23 , conn)
print("question-23 :- \n",question_23)
print("\n" , "="*100 , "\'n")

# 24. INNER JOIN employees and departments
#     Find departments where 
#     ALL employees earn above 50000
#     Hint: GROUP BY + HAVING MIN(salary) > 50000
query_24 = ''' 
    SELECT 
        d.dept_name,
        e.name,
        e.salary
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
    HAVING MIN(e.salary) > 50000
''' 
question_24 = pd.read_sql(query_24 , conn)
print(question_24)
print("\n" , "="*100 , "\'n")

# 25. INNER JOIN employees and departments
#     Find employees who earn MORE than
#     the average salary of their department
#     Hint:
#     JOIN with subquery:
#     JOIN (SELECT dept_id, AVG(salary) 
#           as avg_sal FROM employees 
#           GROUP BY dept_id) avg_table
#     ON e.dept_id = avg_table.dept_id
#     WHERE e.salary > avg_table.avg_sal
# query_25 = ''' 
#     SELECT
#         e.name
#     FROM employees e
#     JOIN 
#         (SELECT 
#             d.dept_id, 
#             AVG(e.salary) as avg_sal 
#             FROM employees e
#             GROUP BY d.dept_id) as avg_table
#     ON e.dept_id = avg_table.dept_id
#     WHERE e.salary > avg_table.avg_sal
# ''' 
# question_25 = pd.read_sql(query_25 ,conn)
# print(question_25)


# 26. INNER JOIN employees and salary_history
#     Find year wise total salary 
#     cost for the company
#     Hint: GROUP BY year ORDER BY year
query_26 = ''' 
    SELECT 
        s.year,
        SUM(e.salary) as total_salary
    FROM employees e
    INNER JOIN salary_history s
        ON e.emp_id = s.emp_id
    GROUP BY s.year
    ORDER BY s.year DESC
''' 
question_26 = pd.read_sql(query_26 , conn)
print(question_26)
print("\n" , "="*100 , "\'n")



# 27. INNER JOIN employees and departments
#     Create a full employee report:
#     name, dept_name, location, 
#     manager, salary
#     Only Technology and Finance departments
#     ORDER BY dept_name, salary DESC
query_27 = ''' 
    SELECT
        e.name,
        d.dept_name,
        d.location,
        d.manager,
        e.salary
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    WHERE d.dept_name IN ("Technology","Finance")
    GROUP BY e.name , d.dept_name
    ORDER BY d.dept_name , e.salary DESC
''' 
question_27 = pd.read_sql(query_27,conn)    
print(question_27)
print("\n" , "="*100 , "\'n")


# 28. LEFT JOIN employees and departments
#     Find which departments have 
#     NO employees at all
#     Hint: 
#     FROM departments d
#     LEFT JOIN employees e ON...
#     WHERE e.emp_id IS NULL

query_28 = ''' 
    SELECT
        d.dept_name,
        e.name
    FROM employees e
    RIGHT JOIN departments d
        ON e.dept_id =  d.dept_id 
    WHERE e.emp_id IS NULL
    GROUP BY d.dept_name
''' 
question_28 = pd.read_sql(query_28 , conn)
print(question_28)
print("\n" , "="*100 , "\'n")



# 29. INNER JOIN all 3 tables
#     Find top 3 employees by 
#     total earnings (salary + bonus)
#     across all years
#     Hint: SUM(s.salary + s.bonus)
#     GROUP BY name ORDER BY DESC LIMIT 3
query_29 = '''
    SELECT
        e.name,
        *,
        SUM(s.salary + s.bonus) as total_earnings
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    INNER JOIN salary_history s
        ON e.emp_id = s.emp_id
    GROUP BY e.name
    ORDER BY total_earnings DESC
    LIMIT 3
''' 
question_29 = pd.read_sql(query_29,conn)
print(question_29)
print("\n" , "="*100 , "\'n")



# 30. INNER JOIN employees and departments
#     Find the complete department summary:
#     dept_name, location, manager,
#     emp_count, avg_salary, 
#     total_salary_cost, max_salary
#     All in ONE query
#     This is a real interview question!
query_30 = ''' 
    SELECT 
        d.dept_name,
        d.location,
        d.manager,
        COUNT(*) as emp_count,
        AVG(e.salary) as avg_sal,
        SUM(e.salary) as total_salary_cost,
        MAX(e.salary) as max_salary
    FROM employees e
    INNER JOIN departments d
        ON e.dept_id = d.dept_id
    GROUP BY d.dept_name
'''
question_30 = pd.read_sql(query_30 , conn)
print(question_30)
```
