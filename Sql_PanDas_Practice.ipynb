```python
import pandas as pb
import sqlite3

conn = sqlite3.connect("database_2.db")

cursor = conn.cursor()
```

```python
cursor.execute(''' 
    CREATE TABLE customers_2 (
        customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        email TEXT UNIQUE NOT NULL,
        city TEXT,
        signup_date DATE 
    )
 ''')
```

```python
cursor.execute(''' 
    CREATE TABLE products (
        product_id INTEGER PRIMARY KEY AUTOINCREMENT,
        product TEXT NOT NULL,
        category TEXT,
        price REAL NOT NULL,
        stock INTEGER DEFAULT 0
    ) 
 ''')
```

```python
# creating the review data table

cursor.execute(''' 
CREATE TABLE IF NOT EXISTS reviews (
    review_id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER NOT NULL,
    customer_id INTEGER NOT NULL,
    rating INTEGER CHECK(rating >=1 AND rating <=5),
    review_text TEXT,
    review_date DATE,
    FOREIGN KEY (product_id) REFERENCES products(product_id),
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
)
 ''')
```

```python
# Question 1
# Select the columns:
# sex, age, fare
# from the Titanic table.
query ='''
    SELECT sex , age, fare
    FROM titanic
'''

columns_filter = pd.read_sql(query,conn)
print(columns_filter)
print("\n", "="*100,"\n")


# Question 2
# Select the columns:
# sex, age, class
# for all passengers.
query = '''
    SELECT sex ,age, class
    FROM titanic
'''
columns_filter_2 = pd.read_sql(query,conn)
print(columns_filter_2)
print("\n", "="*100,"\n")

# -------------------------------------------------------------------------------

# 🚀 LEVEL 2 — WHERE Filtering

# Question 3
# Find all female passengers.
# Columns:
# sex, age, class
query = ''' 
    SELECT sex,age, class
    FROM titanic
    WHERE sex = "female"
'''
female_pass = pd.read_sql(query,conn)
print(female_pass)
print("\n", "="*100,"\n")


# Question 4
# Find passengers whose age is greater than 60.
# Columns:
# sex, age, class
query = '''
    SELECT sex, age, class
    FROM titanic
    WHERE age > 60
'''
age_60 = pd.read_sql(query,conn)
print(age_60)
print("\n", "="*100,"\n")


# Question 5
# Find passengers who did not survive.
# Columns:
# sex, age, class, survived
query='''
    SELECT sex, age, class, survived
    FROM  titanic
    WHERE  survived = 0
'''

survived_pass = pd.read_sql(query,conn)
print(survived_pass)
print("\n", "="*100,"\n")

# -------------------------------------------------------------------------------


# 🚀 LEVEL 3 — Multiple Conditions

# Question 6
# Find female passengers in first class.
# Columns:
# sex, age, class, fare
query = '''
    SELECT sex , age, class, fare
    FROM titanic
    WHERE sex = 'female' AND class = 'First'
'''
female_first = pd.read_sql(query ,conn)
print(female_first)
print("\n", "="*100,"\n")


# Question 7
# Find passengers who:
# age > 50 AND fare > 100
# Columns:
# sex, age, fare, class
query= '''
    SELECT sex, age, fare, class
    FROM titanic
    WHERE  age>50 AND fare>100
'''
pass_age_fare = pd.read_sql(query, conn)
print(pass_age_fare)
print("\n", "="*100,"\n")


# Question 8
# Find passengers who embarked from:
# C or Q
# Columns:
# sex, age, embarked
query = '''
    SELECT sex, age, embarked
    FROM titanic
    WHERE  embarked IN ("C", "Q")
'''
embarked_C_Q = pd.read_sql(query , conn)
print(embarked_C_Q)
print("\n", "="*100,"\n")


# Question 9
# Find passengers whose fare is between 50 and 100.
# Columns:
# sex, fare, class
query = '''
    SELECT sex, fare, class
    FROM titanic
    WHERE fare BETWEEN 50 AND 100
'''
pass_fare_filter = pd.read_sql(query ,conn)
print(pass_fare_filter)
print("\n", "="*100,"\n")

# -------------------------------------------------------------------------------


# 🚀 LEVEL 4 — ORDER BY

# Question 10
# Show 10 passengers who paid the highest fare.
# Columns:
# sex, fare, class

query = '''
    SELECT sex, fare, class
    FROM titanic
    ORDER BY fare DESC LIMIT 10
'''
high_fare = pd.read_sql(query,conn)
print(high_fare)
print("\n", "="*100,"\n")


# Question 11
# Show the 10 youngest passengers.
# Columns
# sex, age, class
query = '''
    SELECT sex, age, class
    FROM titanic
    ORDER BY age LIMIT 10
'''
young_pass = pd.read_sql(query, conn)
print(young_pass)
print("\n", "="*100,"\n")

filter_age = titanic[['sex', 'age', 'class']].sort_values(by="age",ascending=True).head(10)
print(filter_age)
print("\n", "="*100,"\n")

# -------------------------------------------------------------------------------

# 🚀 LEVEL 5 — Aggregations

# Question 12
# Count the total number of passengers.
query = '''
    SELECT COUNT(*)
    FROM titanic
'''
total_pass_tit = pd.read_sql(query,conn)
print(total_pass_tit)
print("\n", "="*100,"\n")

# Question 13
# Find the average age of passengers.
query = '''
    SELECT AVG(age)
    FROM titanic
'''
avg_age_pass = pd.read_sql(query,conn)
print(avg_age_pass)
print("\n", "="*100,"\n")

# Question 14
# Find the maximum fare paid.
query = '''
    SELECT MAX(fare)
    FROM titanic
'''
max_fare = pd.read_sql(query, conn)
print(max_fare)
print("\n", "="*100,"\n")

# -------------------------------------------------------------------------------

# 🚀 LEVEL 6 — GROUP BY (Important)

# Question 15
# Count how many passengers are in each class.
# Columns
# class
# passenger_count
query = '''
    SELECT class , COUNT(*) AS passenger_count
    FROM titanic
    GROUP BY class;
'''
pass_class = pd.read_sql(query,conn)
print(pass_class)
print("\n", "="*100,"\n")

# Question 16
# Find the average fare per class.
# Columns:
# class
# avg_fare
query = '''
    SELECT class, AVG(fare) AS avg_fare
    FROM titanic
    GROUP BY class;
'''
avg_fare_class = pd.read_sql(query,conn)
print(avg_fare_class)
print("\n", "="*100,"\n")

# Question 17
# Count how many males and females are there.
# Columns:
# sex
# passenger_count
query = '''
    SELECT sex , COUNT(sex)
    FROM titanic
    GROUP BY sex
'''
count_males_female = pd.read_sql(query, conn)
print(count_males_female)
print("\n", "="*100,"\n")

# Question 18
# Find the average age of passengers by sex.
# Columns:
# sex
# avg_age
query = '''
    SELECT sex , AVG(age) AS avg_age
    FROM titanic
    GROUP BY sex
'''
avg_age_sex = pd.read_sql(query,conn)
print(avg_age_sex)
print("\n", "="*100,"\n")

# -------------------------------------------------------------------------------

# 🚀 LEVEL 7 — Analytical Thinking

# Question 19
# Find how many passengers survived in each class.
# Columns:
# class
# survived
# passenger_count
query = '''
    SELECT class , survived, COUNT(*)
    FROM titanic
    GROUP BY class
'''
survived_class = pd.read_sql(query,conn)
print(survived_class)
print("\n", "="*100,"\n")

# Question 20
# Find the maximum fare paid by passengers from each embark_town.
# Columns:
# embark_town
# max_fare
query = '''
    SELECT embark_town , MAX(fare) AS max_fare
    FROM titanic
    GROUP BY embark_town
'''
max_fare_pass = pd.read_sql(query, conn)
print(max_fare_pass)
print("\n", "="*100,"\n")
```
