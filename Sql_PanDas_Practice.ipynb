```python
import pandas as pandas
import numpy as numpy
import sqlite3

print("SQL Version is : - ", sqlite3.version)

connection = sqlite3.connect("company.db")

cursor = connection.cursor()

cursor.execute("""
CREATE TABLE employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER
)
""")

connection.commit()
```

```python
import pandas as pd
import sqlite3
import seaborn as sns

# Create a new database
conn = sqlite3.connect('data_analysis.db')

# Load datasets
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')  # Restaurant tips dataset

# Save to database
titanic.to_sql('titanic', conn, if_exists='replace', index=False)
iris.to_sql('iris', conn, if_exists='replace', index=False)
tips.to_sql('tips', conn, if_exists='replace', index=False)

print("Database created successfully!")
print(f"Titanic: {len(titanic)} rows")
print(f"Iris: {len(iris)} rows")
print(f"Tips: {len(tips)} rows")

conn.close()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('data_analysis.db')

# Query to get all table names
query = "SELECT name FROM sqlite_master WHERE type='table'"
tables = pd.read_sql(query, conn)
print(tables)

conn.close()
```

```python
conn = sqlite3.connect('data_analysis.db')

# Get column information
query = "PRAGMA table_info(titanic)"
columns_info = pd.read_sql(query, conn)
print(columns_info[['name', 'type']])

conn.close()
```
