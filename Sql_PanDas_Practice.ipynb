# Database_2 Created and Practiced with Inserting data

```python
import pandas as pb
import sqlite3

conn = sqlite3.connect("database_2.db")

cursor = conn.cursor()
```

```python
cursor.execute(''' 
    CREATE TABLE customers_2 (
        customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        email TEXT UNIQUE NOT NULL,
        city TEXT,
        signup_date DATE 
    )
 ''')
```

```python
cursor.execute(''' 
    CREATE TABLE products (
        product_id INTEGER PRIMARY KEY AUTOINCREMENT,
        product TEXT NOT NULL,
        category TEXT,
        price REAL NOT NULL,
        stock INTEGER DEFAULT 0
    ) 
 ''')
```

```python
# creating the review data table

cursor.execute(''' 
CREATE TABLE IF NOT EXISTS reviews (
    review_id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER NOT NULL,
    customer_id INTEGER NOT NULL,
    rating INTEGER CHECK(rating >=1 AND rating <=5),
    review_text TEXT,
    review_date DATE,
    FOREIGN KEY (product_id) REFERENCES products(product_id),
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
)
 ''')
```

```python
conn.commit()
print("Table created Successfully")
```

```python
# SECTION A — Basic SELECT (Warm Up)

# 1.  Select all columns from titanic table
#     (first 5 rows only using LIMIT)
query = '''
    SELECT * 
    FROM  titanic 
    LIMIT 5
'''
all_column = pd.read_sql(query ,conn)
print(all_column)
print("\n", "="*100,"\n")

# 2.  Select only name, sex, age, fare columns
#     Wait — titanic has no name column.
#     Select: survived, pclass, sex, age, fare
#     (first 10 rows)
query = '''
    SELECT survived , pclass , sex , age , fare
    FROM titanic
    LIMIT 10
'''
cust_colm = pd.read_sql(query,conn)
print(cust_colm)
print("\n", "="*100,"\n")

# 3.  Select all DISTINCT values in "sex" column
#     Hint: SELECT DISTINCT sex FROM titanic
query = '''
    SELECT DISTINCT sex
    FROM titanic
'''
sex_dist = pd.read_sql(query, conn)
print(sex_dist)
print("\n", "="*100,"\n")

# 4.  Select all DISTINCT values in "class" column
query = '''
    SELECT DISTINCT class
    FROM  titanic
'''
class_unq = pd.read_sql(query,conn)
print(class_unq)
print("\n", "="*100,"\n")

# 5.  Select all DISTINCT values in "embarked" column
query = '''
    SELECT DISTINCT embarked
    FROM titanic
'''
emabrk_unq = pd.read_sql(query,conn)
print(emabrk_unq)
print("\n", "="*100,"\n")

# 6.  Count total number of passengers
#     Hint: SELECT COUNT(*) FROM titanic
query = ''' 
    SELECT COUNT(*)
    FROM  titanic
''' 
total_pass = pd.read_sql(query,conn)
print(total_pass)
print("\n", "="*100,"\n")

# 7.  Count total number of passengers 
#     who survived
#     Hint: WHERE survived = 1
query = '''
    SELECT COUNT(*)
    FROM titanic
    WHERE survived = 1
''' 
survived_peop = pd.read_sql(query,conn)
print(survived_peop)
print("\n", "="*100,"\n")


# 8.  Count total number of passengers 
#     who did NOT survive
#     Hint: WHERE survived = 0
query = '''
    SELECT COUNT(*)
    FROM titanic
    WHERE survived = 0
'''
died_peop = pd.read_sql(query,conn)
print(died_peop)
print("\n", "="*100,"\n")

# # -----------------------------------------------
```

```python
# SECTION B — WHERE Clause (Filtering)

# 9.  Get all passengers who are female
query = '''
    SELECT *
    FROM titanic
    WHERE sex = 'female'
''' 
female_pass = pd.read_sql(query,conn)
print(female_pass)
print("\n", "="*100,"\n")

# 10. Get all passengers from First class
#     Hint: WHERE pclass = 1
query = ''' 
    SELECT *
    FROM titanic
    WHERE pclass=1
''' 
first_class_pass = pd.read_sql(query,conn)
print(first_class_pass)
print("\n", "="*100,"\n")

# 11. Get all passengers whose age is above 50
query = ''' 
    SELECT *
    FROM titanic
    WHERE age > 50
''' 
age_abov_50 = pd.read_sql(query,conn)
print(age_abov_50)
print("\n", "="*100,"\n")

# 12. Get all passengers who survived 
#     AND are female
query = ''' 
    SELECT *
    FROM titanic
    WHERE   survived = 1 AND sex='female' 
''' 
survive_fem = pd.read_sql(query,conn)
print(survive_fem)
print("\n", "="*100,"\n")

# 13. Get all passengers who did NOT survive
#     AND are male
query = ''' 
    SELECT *
    FROM titanic
    WHERE survived = 0 AND sex = "male"
''' 
male_sur = pd.read_sql(query,conn)
print(male_sur)
print("\n", "="*100,"\n")

# 14. Get all passengers whose fare 
#     is above 100
query = ''' 
    SELECT *
    FROM titanic
    WHERE fare > 100
''' 
fare_100 = pd.read_sql(query,conn)
print(fare_100)
print("\n", "="*100,"\n")

# 15. Get all passengers from class 1 
#     OR class 2
#     Hint: WHERE pclass IN (1, 2)
query = ''' 
    SELECT *
    FROM titanic
    WHERE pclass IN (1,2)
''' 
class_1_2 = pd.read_sql(query,conn)
print(class_1_2)
print("\n", "="*100,"\n")

# 16. Get all passengers whose age 
#     is between 18 and 35
#     Hint: WHERE age BETWEEN 18 AND 35
query = ''' 
    SELECT *
    FROM titanic
    WHERE age BETWEEN 18 AND 35
''' 
age_filter = pd.read_sql(query,conn)
print(age_filter)
print("\n", "="*100,"\n")

# 17. Get all passengers who embarked 
#     from 'S' (Southampton)
query = ''' 
    SELECT *
    FROM titanic
    WHERE embarked = "S"
''' 
embark_S = pd.read_sql(query,conn)
print(embark_S)
print("\n", "="*100,"\n")

# 18. Get passengers where deck 
#     is NOT NULL
#     Hint: WHERE deck IS NOT NULL
query = ''' 
    SELECT *
    FROM titanic
    WHERE deck IS NOT NULL
''' 
deck_notnull = pd.read_sql(query,conn)
print(deck_notnull)
print("\n", "="*100,"\n")

# # -----------------------------------------------
```
